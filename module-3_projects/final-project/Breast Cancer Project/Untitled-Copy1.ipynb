```python
import pandas as pd
import glob
import os
```

```python
os.path.basename('./CANCER INCIDENCE IN FIVE CONTINENTS VOLUME XI/557800099.csv').split('.')[0]
```

```python
df_seed = pd.read_csv('./CANCER INCIDENCE IN FIVE CONTINENTS VOLUME XI/557800099.csv', header=None)
df_seed['filename'] = os.path.basename('./CANCER INCIDENCE IN FIVE CONTINENTS VOLUME XI/557800099.csv').split('.')[0]
df_seed
```

```python
df_seed = df_seed.rename(columns={0:'sex',
                       1:'site',
                       2:'age',
                       3:'number',
                       4:'years_at_risk'})
```

```python
path = './CANCER INCIDENCE IN FIVE CONTINENTS VOLUME XI/'
all_files = glob.glob(path + "/*.csv")
for filename in all_files:
    df_file = pd.read_csv(filename, header=None)
    df_file['filename'] = os.path.basename(filename).split('.')[0]
    df_file = df_file.rename(columns={0:'sex',
                       1:'site',
                       2:'age',
                       3:'number',
                       4:'years_at_risk'})
    df_seed=pd.concat([df_seed,df_file])
```

```python
df_seed
```

```python
data = pd.read_fwf("./CANCER INCIDENCE IN FIVE CONTINENTS VOLUME XI/cancer_detailed.txt")
```

```python
data = data.drop(columns=['Unnamed: 2', 'Unnamed: 3']).dropna(0)
```

```python
data = data.rename(columns={'001': 'code', 'All sites (C00-96)' : 'cancer'})
```

```python
data['cancer'] = data['cancer'].str.split('(', expand=True, n=0)
```

```python
data['cancer']  = data['cancer'].str.strip()
```

```python
cancer_dict = dict(zip(data.code, data.cancer))
cancer_dict
```

```python
data
```

```python
df_seed['number'] = df_seed['number'].map(cancer_dict)
```

```python
df_seed
```

```python
country =  pd.read_fwf("./CANCER INCIDENCE IN FIVE CONTINENTS VOLUME XI/registry_detailed.txt", header=None)
```

```python
country_df = country.drop(columns=2)
country_df[1] = country_df[1].str.strip('*').str.split(',', expand=True, n=0)
country_df = country_df.rename(columns={0:'code', 1:'country'})
country_df['code'] = country_df['code'].astype(str)
country_dict = dict(zip(country_df.code, country_df.country))
```

```python
df = df_seed.copy()
df['filename'] = df['filename'].map(country_dict)
df = df.rename(columns = {'filename':'country'})
```

```python
df = df.rename(columns={'number': 'cancer'})
```
