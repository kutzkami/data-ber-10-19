```python
import pandas as pd
import glob
import os
```

```python
df_seed = pd.read_csv('./CI5plus_Summary_April2019/Europe_Cases.csv')
```

```python
df_seed
```

```python
path = './CI5plus_Summary_April2019/'
all_files = glob.glob(path + "/*_Cases.csv")
for filename in all_files:
    df_file = pd.read_csv(filename)
    df_seed=pd.concat([df_seed,df_file])
```

```python
df = df_seed.copy()
```

```python
df = df[['REGISTRY','YEAR','SEX','CANCER','TOTAL']]
```

```python
cancer = pd.read_csv("./CI5plus_Summary_April2019/cancer_summary.csv", header=None)
cancer = cancer.rename(columns={0:'code',1:'cancer'})
cancer['cancer'] = cancer['cancer'].str.split('(',expand=True, n=0)
cancer['cancer'] = cancer['cancer'].str.strip()
df['CANCER'] = df['CANCER'].map(dict(zip(cancer.code, cancer.cancer)))
```

```python
df.dtypes
```

```python
country_seed = pd.read_csv('./CI5plus_Summary_April2019/Europe_r.csv')
country_seed['CONTINENT'] = os.path.basename('./CI5plus_Summary_April2019/Europe_r.csv').split('_')[0]
```

```python
path = './CI5plus_Summary_April2019/'
all_files = glob.glob(path + "/*_r.csv")
for filename in all_files:
    df_file = pd.read_csv(filename)
    df_file['CONTINENT'] = os.path.basename(filename).split('_')[0]
    country_seed=pd.concat([country_seed,df_file])
```

```python
"""df.to_csv('cancer_timeline.csv')"""
country_seed
```

```python
country_seed['NAME'] = country_seed['NAME'].str.replace('New_Zealand', 'New Zealand')
country_seed['NAME'] = country_seed['NAME'].str.replace('Western_Australia', 'Australia')
country_seed['NAME'] = country_seed['NAME'].str.replace('Costa_Rica', 'Costa Rica')
country_seed['NAME'] = country_seed['NAME'].str.replace('South_Australia', 'South Australia')
```

```python
country_seed['NAME'] = country_seed['NAME'].str.split('_', expand=True)
country_seed['NAME'] = country_seed['NAME'].str.split('-', expand=True)

country_seed
```

```python
country = country_seed[['REGISTRY', 'NAME']]
country_dict = dict(zip(country.REGISTRY,country.NAME))
```

```python
df['REGISTRY'] = df['REGISTRY'].map(country_dict)
```

```python
df['REGISTRY'].value_counts()
```

```python
df = df.drop_duplicates()
df.to_csv('cancer_timeline.csv')
```
