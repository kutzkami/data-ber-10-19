# TOTAL CANCER DEATHS BY TYPE

```python
df = pd.read_csv('./OUR_WORLD_IN_DATA/total-cancer-deaths-by-type.csv')
```

```python
df.columns = [i.split('cancer')[0].strip() for i in df.columns.to_list()]
df.columns = [i.split('(')[0].strip() for i in df.columns.to_list()]
```

```python
cancer = df.columns.to_list()[3:]
df1 = df[['Entity','Code','Year','Liver']]
df1['Cancer'] = df1.columns.to_list()[3]
df1 = df1.rename(columns={df1.columns.to_list()[3]:'Death'})
```

```python
df_final = df1.copy()
for i in cancer:
    df2 = df[['Entity','Code','Year',i]]
    df2['Cancer'] = df2.columns.to_list()[3]
    df2 = df2.rename(columns={df2.columns.to_list()[3]:'Death'})
    df_final=df_final.append(df2, ignore_index=True)
```

```python
df_final
```

```python
df_final = df_final.drop_duplicates()
```

```python
df_final
```

```python
df_final[df_final.isna().any(axis=1)]
```

```python
df_final.to_csv("total-cancer-deaths-by-type-clean.csv")
```

# share of population by cancer types
When broken down by type of cancer we see that breast cancer is globally the most prevalent form. Following breast cancer are prostrate and colon & rectum cancer.

When you change this chart to another country you see that across most countries these are the top three cancer forms.

```python
df = pd.read_csv('./OUR_WORLD_IN_DATA/share-of-population-with-cancer-types.csv')
```

```python
df.columns = [i.split('cancer')[0].strip() for i in df.columns.to_list()]
df.columns = [i.split('&')[0].strip() for i in df.columns.to_list()]
df.columns = [i.split('and')[0].strip() for i in df.columns.to_list()]
df.columns = [i.split(',')[0].strip() for i in df.columns.to_list()]
df = df.rename(columns={'Tracheal': 'Lung', 'Non-melanoma skin': 'Skin'})
```

```python
cancer = df.columns.to_list()[3:]
df['Top']  = df[cancer].idxmax(axis=1)
```

```python
df1 = df[['Entity','Code','Year','Top','Liver']]
df1['Cancer'] = df1.columns.to_list()[4]
df1 = df1.rename(columns={df1.columns.to_list()[4]:'Death'})
```

```python
df_final = df1.copy()
for i in cancer:
    df2 = df[['Entity','Code','Year','Top',i]]
    df2['Cancer'] = df2.columns.to_list()[4]
    df2 = df2.rename(columns={df2.columns.to_list()[4]:'Death'})
    df_final=df_final.append(df2, ignore_index=True)
```

```python
df_final = df_final.drop_duplicates()
```

```python
df_final.to_csv('share-of-population-with-cancer-types-clean.csv')
```
