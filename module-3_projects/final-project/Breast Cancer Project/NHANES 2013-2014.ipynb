# Medical Conditions (MCQ_H)
SEQN - Respondent sequence number
- MCQ230a - What kind of cancer (Code 14 is Breast, many codes many different cancers)

- MCQ240e - Age when breast cancer first diagnosed

##### choosing not diagnosed ppl
- MCQ220 - Ever told you had cancer or malignancy
-no = 2

```python
df_med = pd.read_sas('./NHANES 2013-2014/MCQ_H.XPT')
```

```python
df_med.head()
```

```python
# match only dose SEQN matching the one inf df_demo (seqn variable)
df_med = df_med.loc[df_med['SEQN'].isin(df_demo['SEQN'])]
```

```python
# is the length the same?
len(df_demo), len(df_med)
```

```python
# MCQ230A - What kind of cancer (Code 14 is Breast, many codes many different cancers)
# MCQ240E - Age when breast cancer first diagnosed
# MCQ220 - Ever told you had cancer or malignancy === no = 2

df_med = df_med[['SEQN', 'MCQ230A', 'MCQ240E', 'MCQ220']]
```

```python
# add column breast cancer 'Diagnosis' 1 no 0
df_med['Breast Cancer'] = np.where(df_med['MCQ230A'] == 14,1,0)
```

```python
df_med['Breast Cancer'].value_counts()
```

```python
# create a column cancer occurance 'Cancer' 1, no 0
df_med['Diagnosed'] = np.where(df_med['MCQ220'] == 2,0,1)
```

```python
df_med['Diagnosed'].value_counts()
```

```python
# subset to have women that were diagnosed with breast cancer and women that were not diagnosed with any cancer
df_med = df_med[(df_med['Breast Cancer'] == 1) | (df_med['Diagnosed'] == 0)]
```

```python
# changing the column name to diagnosis age and saving the dataset
df_med = df_med.rename(columns={'MCQ240E' : 'Diagnosis Age'})[['SEQN', 'Diagnosis Age', 'Breast Cancer', 'Diagnosed']]
```

```python
# merge df_med with df_demo
df = pd.merge(df_demo, df_med, on='SEQN', how='inner')
len(df)
```

# Clean up

```python
df.info()
```

```python
# There are some missing values all around, see how the df looks when you remove them
df_copy = df.copy()
```

```python
# first drop diagnosis age since it only reflects the subset of women with breast cancer
df_copy = df_copy.drop(columns='Diagnosis Age')
```

```python
# get rid off the missing valuess
df_copy = df_copy.dropna()
df_copy
# there were some women that were diagnosed with breast cancer that were not completely analysed. 
```

```python
df = pd.merge(df_copy, df[['SEQN', 'Diagnosis Age']], how='inner', on=['SEQN'])
```

#### Clean the column names, they ugly bro

```python
df.columns = df.columns.str.lower().str.capitalize()
```

```python
#rename the Seqn column
df = df.rename(columns={'Seqn':'Id'})
```

```python
# control for duplicates
df.duplicated().sum()
```

```python
# control for missing values
df.isna().sum()
```

In this dataset, 78 women have been diagnosed with cancer. I will randomly choose 78 healthy women to have groups of comparable sizes

```python
df_healthy = df.loc[df['Breast cancer'] == 0]
df_healthy = df_healthy.sample(n=78)
```

```python
df_final = df_healthy.append(df[df['Breast cancer']==1])
```

```python
df_final = df_final.sort_values(by=['Id'])
```
